### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
condition1 = (data['country']=='Lithuania') | (data['country']=='Latvia') | (data['country']=='Estonia')
condition2 = (data['category']==4) | (data['category']==12) | (data['category']==21)
condition3 = (data['year']>=2005) & (data['year']<=2010)
condition4 = (data['quantity']>0)
```

```python
data_filter = data[condition1 & condition2 & condition3 & condition4]
data_filter
```

```python
sum(data_filter['country'].value_counts().values)
```
